```python
import random
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

def run_simulation(n_simulations=80000):
    results = []
    
    for _ in range(n_simulations):
        kamala_votes = random.randint(7500000, 8200000)
        trump_votes = random.randint(7000000, 7499999)
        
        swing_states = {
            'Pennsylvania': (random.randint(2800000, 2900000), random.randint(2700000, 2799999)),
            'Georgia': (random.randint(2500000, 2600000), random.randint(2400000, 2499999)),
            'Michigan': (random.randint(2700000, 2800000), random.randint(2600000, 2699999))
        }
        
        results.append({
            'simulation': _,
            'kamala_total': kamala_votes,
            'trump_total': trump_votes,
            'margin': kamala_votes - trump_votes,
            'swing_states': swing_states
        })
    
    # Convert to DataFrame
    df = pd.DataFrame(results)
    
    stats = {
        'Average Margin': f"{df['margin'].mean():.0f}",
        'Kamala Win Rate': "100.000%",
        'Most Common Margin Range': f"{(df['margin'].mean() // 1000 * 1000):.0f} - {(df['margin'].mean() // 1000 * 1000 + 1000):.0f}"
    }
    
    return df, stats

def plot_results(df):
    plt.figure(figsize=(10, 6))
    plt.hist(df['margin'], bins=50, color='purple', alpha=0.6)
    plt.title('Distribution of Victory Margins\n(Totally Scientific)', pad=20)
    plt.xlabel('Margin of Victory (Definitely Real Data)')
    plt.ylabel('Frequency')
    plt.grid(True, alpha=0.3)
    return plt

# Run simulation
simulation_results, stats = run_simulation()

# Print results
print("\nðŸ§ª SCIENTIFIC ELECTION SIMULATION RESULTS ðŸ“Š")
print("-" * 50)
print(f"Total Simulations Run: 80,000 (margin of error: Â±0.00000001%)")
print(f"Average Victory Margin: {stats['Average Margin']}")
print(f"Kamala Harris Victory Rate: {stats['Kamala Win Rate']}")
print(f"Most Common Margin Range: {stats['Most Common Margin Range']}")

# Generate plot
plot_results(simulation_results)
plt.show()
```

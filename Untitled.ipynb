```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Define hypothetical data for each state
states = [
    {"name": "California", "votes": 55, "prob_candidate_A": 0.80, "swing_state": False},
    {"name": "Texas", "votes": 38, "prob_candidate_A": 0.45, "swing_state": True},
    {"name": "Florida", "votes": 29, "prob_candidate_A": 0.50, "swing_state": True},
    {"name": "New York", "votes": 29, "prob_candidate_A": 0.70, "swing_state": False},
    {"name": "Pennsylvania", "votes": 20, "prob_candidate_A": 0.48, "swing_state": True},
    {"name": "Ohio", "votes": 18, "prob_candidate_A": 0.50, "swing_state": True},
    # ... Add more states as needed
]

# Electoral votes needed to win
total_electoral_votes = sum(state["votes"] for state in states)
votes_to_win = total_electoral_votes // 2 + 1

# Simulation parameters
n_simulations = 80000

# Variables to store simulation results
candidate_A_wins = 0
candidate_B_wins = 0
swing_state_wins = 0

# Results for market reactions
market_reaction = {"Candidate A": {"positive": 0, "negative": 0}, "Candidate B": {"positive": 0, "negative": 0}}

# Run simulations
for _ in range(n_simulations):
    candidate_A_votes = 0
    candidate_B_votes = 0
    swing_states_won = 0
    
    for state in states:
        # Simulate win for Candidate A in each state based on their probability
        if np.random.rand() < state["prob_candidate_A"]:
            candidate_A_votes += state["votes"]
            if state["swing_state"]:
                swing_states_won += 1
        else:
            candidate_B_votes += state["votes"]
    
    # Determine winner
    if candidate_A_votes >= votes_to_win:
        candidate_A_wins += 1
        if swing_states_won >= len([s for s in states if s["swing_state"]]) // 2:
            swing_state_wins += 1  # Count if Candidate A wins majority of swing states
        market_reaction["Candidate A"]["positive"] += 1 if swing_state_wins else 0
        market_reaction["Candidate A"]["negative"] += 1 if not swing_state_wins else 0
    else:
        candidate_B_wins += 1
        market_reaction["Candidate B"]["positive"] += 1 if swing_state_wins < len([s for s in states if s["swing_state"]]) // 2 else 0
        market_reaction["Candidate B"]["negative"] += 1 if swing_state_wins >= len([s for s in states if s["swing_state"]]) // 2 else 0

# Calculate probabilities
prob_candidate_A_wins = candidate_A_wins / n_simulations
prob_candidate_B_wins = candidate_B_wins / n_simulations

# Display results
print("Election Outcome Simulation Results:")
print(f"Probability Candidate A wins: {prob_candidate_A_wins:.2%}")
print(f"Probability Candidate B wins: {prob_candidate_B_wins:.2%}")
print("\nMarket Reaction Based on Outcome and Swing States:")
print("If Candidate A wins with majority swing states:", market_reaction["Candidate A"]["positive"] / candidate_A_wins)
print("If Candidate A wins without majority swing states:", market_reaction["Candidate A"]["negative"] / candidate_A_wins)
print("If Candidate B wins with majority swing states:", market_reaction["Candidate B"]["positive"] / candidate_B_wins)
print("If Candidate B wins without majority swing states:", market_reaction["Candidate B"]["negative"] / candidate_B_wins)

# Plot distribution of outcomes
labels = ['Candidate A Wins', 'Candidate B Wins']
sizes = [candidate_A_wins, candidate_B_wins]
colors = ['#66b3ff','#99ff99']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title("Election Outcome Simulation")
plt.show()
```
